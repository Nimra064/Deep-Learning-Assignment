### **Step # 7**
MODEL CREAT

```python
from keras import models
from keras import layers
```

```python
network=models.Sequential()
network.add(layers.Dense(34 , activation='relu'  , input_shape=(34,)))
network.add(layers.Dense(16 , activation='relu' ))
network.add(layers.Dense(1 , activation='sigmoid'))
```

```python
learning_rate=0.1
network.compile(optimizer='sgd' ,loss='binary_crossentropy'  , metrics=['accuracy'])
```

```python
history=network.fit(train_data , train_labals ,  epochs=50 , batch_size=1 , validation_split=0.33  )
```

```python
test_loss , test_acc=network.evaluate(test_data , test_labals)
```

### **Step** # **8**
PLOT A GRAPH

```python
# PLOT GRAPG BETWEEN ACCURACY AND EPOCH
acc=history.history['accuracy']
val_acc=history.history['val_accuracy']
epoch=range(1,51)
plt.plot(epoch , acc , 'g' , label='Training_Accuracy')
plt.plot(epoch , val_acc , 'r' , label='Validation_Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
# PLOT GRAPG BETWEEN loss AND EPOCH
los=history.history['loss']
val_los=history.history['val_loss']
epoch=range(1,51)
plt.plot(epoch , los , 'g' , label='Training_Loss')
plt.plot(epoch , val_los , 'r' , label='Validation_Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
